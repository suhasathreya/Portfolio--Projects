```python
#Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\suhasathreya\Downloads\movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(pct_missing)
```

```python
# Drop rows with missing data

df = df.dropna(how='any', axis=0)

#Check for missing values again

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(pct_missing)
    
df.head()
```

```python
# Datatypes for columns

df.dtypes
```

```python
# Change datatype of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
 # Hypothesis 1- Budget has a high correlation with gross
# Hypothesis 2- Company has a correlation
# We will test these now
```

```python
# Scatter plot of budget vs gross revenus

plt.scatter(df['budget'], df['gross'])

plt.title('Budget vs Gross Revenue')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Examining correlation

df.corr(method = 'pearson') # pearson, kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Now checking for correlation between Company and Gross Revenue
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs>0.5]

high_corr
```
